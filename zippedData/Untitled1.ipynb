```python
#  importing data to see highest domestic gross in 5 years
import pandas as pd
df= pd.read_csv('bom.movie_gross.csv.gz')
df.head()
df.drop(['foreign_gross','studio'], axis = 1, inplace=True)
df
```

```python

df.info()
```

```python
#  sorting data in scending year to see the latest year

dff=df.sort_values(by=['year'], ascending = False)
dff
```

```python
# sorting data for latest 4 latest years

four_yrs = dff.loc[(dff['year'] > 2013)
                     & (df['year'] <= 2018)]
four_yrs
```

```python
four_yrs.info()
```

```python
# top performing movies in 4 years
top = four_yrs.sort_values(by=['domestic_gross'], ascending = False)
top

```

```python
# slicing top 20 performing movies thorugh domestic gross collection

top_20 =top.iloc[1:21,:]
top_20
```

```python
# poltting graph to see movies performance

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
fig, ax = plt.subplots()
plt.scatter(top_20['year'], top_20['domestic_gross'])
plt.rcParams['figure.figsize'] = [15,15]
plt.xticks(np.arange(2014, 2019, 1))

ax.set_title('Top 20 movies',fontsize=15)
ax.set_xlabel('Year',fontsize=15)
ax.set_ylabel('Domestic gross in millions',fontsize=15)
```

```python
#  importing titles table to get genres for matching top 20 movies

dfim_basic = pd.read_csv('imdb.title.basics.csv.gz')
dfim_basic.head()
```

```python
# To find top performing movies genre wise
# join table top_20 and imdb.title.basics to get genres of the top_20 movies


basic_yrs = dfim_basic.loc[(dfim_basic['start_year'] > 2013)
                      & (dfim_basic['start_year'] <= 2018)]
dfim_basic.drop(['runtime_minutes','original_title','start_year'], axis =1, inplace =True)

#  renaming "primary_title" to "title" to prepare for merge

dfim_basic.rename(columns ={'primary_title':'title'}, inplace = True)

dfim_basic.set_index('title')
```

```python
#  setting index to prepare for merge

top_20.set_index('title')
```

```python
# cross checking if the Black Panther tconst has come across in merged df

dfim_basic[dfim_basic['title']== 'Black Panther']
```

```python
#  merging top_20 with dfim_basic to get tconst

genre_join = top_20.merge(dfim_basic, how='left')
genre_join
```

```python
#  dropping duplicates from above merge. 
# getting final data for merging with another df to get these genres directors and writers 

gen_final =genre_join.drop_duplicates(['title'], keep ='first')
gen_final
```

```python
# mean of domestic gross by grouping genres  

genre_graph = gen_final.groupby('genres').mean()

genre_graph
```

```python
genre_graph = genre_graph.reset_index()
```

```python
#  plotting bar graph to see which genre is most popular

import matplotlib.pyplot as pltt
x= genre_graph['genres']
y = genre_graph['domestic_gross']
pltt.barh(x,y)
pltt.title('Popular Genres', fontsize=15)
pltt.ylabel('Genre', fontsize=15)
pltt.xlabel('Domestic Gross in trillions', fontsize=15)
pltt.show()
```

```python
#  importing datasets to find out popular movies Directors and Writers
crew = pd.read_csv('imdb.title.crew.csv.gz')
principals = pd.read_csv('imdb.title.principals.csv.gz')
movies = pd.read_csv('imdb.title.basics.csv.gz')
names = pd.read_csv('imdb.name.basics.csv.gz')
```

```python
#  extracting all data related to top 20 movies
example_movies = movies[(movies['primary_title']).isin(['Avengers: Infinity War','Furious 7','Black Panther','Jurassic World','Star Wars: The Last Jedi','Incredibles 2','Rogue One: A Star Wars Story','Beauty and the Beast (2017)','Finding Dory','Avengers: Age of Ultron','Jurassic World: Fallen Kingdom','Wonder Woman','Captain America: Civil War','Jumanji: Welcome to the Jungle','Guardians of the Galaxy Vol. 2','The Secret Life of Pets','The Jungle Book (2016)','Deadpool','Inside Out','American Sniper'])]
example_movies
                     
```

```python
# removing duplicate titles from the dataframe

eg_movies = example_movies[-((example_movies['primary_title']=='Inside Out') & (example_movies['start_year']!=2015))] 
eg_movie = eg_movies[-((eg_movies['primary_title']=='Wonder Woman') & (eg_movies['start_year']!=2017))]
eg_movie
```

```python
# Merging to get directors and writers list

eg_movie = eg_movie.merge(crew, on='tconst')
eg_movie
```

```python
# Merging

director_names = eg_movie.merge(names, left_on='directors', right_on='nconst', how='left')

display(director_names)
director_names.info()
```

```python
#  Dropping duplicate directors

director_list=director_names.dropna(subset=['primary_name'])
director_list
```

```python
#  list of Directors

print(director_list.loc[:,'primary_name'])
```

```python
# Merging to get writers codes

writers_codes = gen_final.merge(principals, left_on='tconst', right_on='tconst', how='left')
display(writers_codes)
```

```python
#  merging to get names of writers from codes

writers=writers_codes.merge(names, left_on='nconst', right_on='nconst', how='left')
display(writers)
```

```python
#  sorting writers from the dataframe

writers_list = writers[writers['category']== 'writer']
writers_list
```
